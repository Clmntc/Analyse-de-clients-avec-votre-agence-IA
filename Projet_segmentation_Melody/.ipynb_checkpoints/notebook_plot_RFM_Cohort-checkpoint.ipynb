```python
#import library

import pandas as pd
import pandas as pd
#viz
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
import squarify

import warnings
```

# Cohort

```python
df = pd.read_csv('data/df_pfm_cohort.csv')
```

```python
df
```

```python
#execute pivot

df_pivot = df.pivot_table(index = 'cohort', columns= 'period_number', values= 'n_customers')
```

```python
cohort_size= df_pivot.iloc[:,0]
retention_matrix = df_pivot.divide(cohort_size, axis=0)
```

```python
#plot heatmap

with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(12, 8), sharey=True, gridspec_kw={'width_ratios': [1, 11]})
    
    # retention matrix
    sns.heatmap(retention_matrix, 
                mask=retention_matrix.isnull(), 
                annot=True, 
                fmt='.0%', 
                cmap='RdYlGn', 
                ax=ax[1])
    ax[1].set_title('Monthly Cohorts: User Retention', fontsize=16)
    ax[1].set(xlabel='# of periods',
              ylabel='')

    # cohort size
    cohort_size_df = pd.DataFrame(cohort_size).rename(columns={0: 'cohort_size'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df, 
                annot=True, 
                cbar=False, 
                fmt='g', 
                cmap=white_cmap, 
                ax=ax[0])

    fig.tight_layout()
```

```python
plt.figure(figsize= (10, 8))

plt.bar(x= 'cohort', height= 'n_customers', width = .7, data=df)

#sizes barticks
plt.xticks(fontsize=12, rotation= 30)

#size titles
plt.xlabel('Date first purchase', fontsize=13, labelpad=20)
plt.ylabel('Customers numbers', fontsize=13, labelpad=20)

plt.show()
```

```python
plt.figure(figsize= (10, 8))

plt.bar(x= 'order_month', height= 'n_customers', width = .7, data=df, color= 'brown')

#sizes barticks
plt.xticks(fontsize=12, rotation= 30)

#size titles
plt.xlabel('Date last purchase', fontsize=13, labelpad=20)
plt.ylabel('Customers numbers', fontsize=13, labelpad=20)

plt.show()
```

# RFM

```python
df_rfm = pd.read_csv('data/df_pfm.csv')
df_rfm
```

```python
plt.figure(figsize= (10, 8))

sns.countplot(df_rfm['Segment'])

#sizes barticks
plt.xticks(fontsize=10, rotation= 30)

#size titles
plt.xlabel('Segments', fontsize=13, labelpad=20)
plt.ylabel('Counts customers', fontsize=13, labelpad=20)

plt.show()
```

```python
plt.figure(figsize= (10, 8))

sns.barplot(x= "Segment", y= 'montant', data=df_rfm)

#sizes barticks
plt.xticks(fontsize=10, rotation= 30)

#size titles
plt.xlabel('Segments', fontsize=13, labelpad=20)
plt.ylabel('Sum spent', fontsize=13, labelpad=20)

plt.show()
```

```python
df_rfm[df_rfm['Segment']== "champions"].value_counts()
```

```python
m = pd.DatetimeIndex(df_rfm['max_date'])
df_rfm['order_month'] = m.to_period('M')
```

```python
df_rfm
```

```python
df_rfm = df_rfm.sort_values("order_month", axis=0, ascending=True)
df_rfm
```

```python
plt.figure(figsize= (20, 10))

sns.barplot(x= 'order_month', y='montant',data=df_rfm, hue='Segment')

#sizes barticks
plt.xticks(fontsize=15, rotation= 30)
#size titles
plt.xlabel('Date last purchase', fontsize=15, labelpad=20)
plt.ylabel('Sum spent', fontsize=15, labelpad=20)

plt.show()
```
